```python
# Temperaturmittelwerte für 1880 und 2024 je Region berechnen
Temperaturmittel =  (
    temp_hist[temp_hist['Jahr'].between(1881, 1886) | temp_hist['Jahr'].between(2019, 2024)]
    .assign(Periode=lambda df: np.where(df['Jahr'].between(1881, 1886), '1881-1886', '2019-2024'))
    .groupby(['Region', 'Periode'])['Temperatur']
    .mean()
    .unstack()
    .reset_index()
    .rename(columns={'1881-1886': 'Mittelwert_1881_1886', '2019-2024': 'Mittelwert_2019_2024'})
)
```

```python
Temperaturmittel['Unterschiede_temp'] = Temperaturmittel['Mittelwert_2019_2024'] - Temperaturmittel['Mittelwert_1881_1886']
Temperaturmittel['Prozentuale Änderung'] = (
    (Temperaturmittel['Unterschiede_temp'] / Temperaturmittel['Mittelwert_1881_1886']) * 100
)
Temperaturmittel.sort_values(by = 'Prozentuale Änderung', ascending=False, inplace=True)
Temperaturmittel
```

```python
window_size = 60  # z.B. gleitender Mittelwert über 60 Monate

# Daten für Deutschland
temp_de = temp_hist[temp_hist['Region'] == 'Deutschland'].sort_values(['Jahr', 'Monat'])
temp_de['Datum'] = pd.to_datetime(temp_de['Jahr'].astype(str) + '-' + temp_de['Monat'].astype(str) + '-01')
temp_de['Rolling_Avg'] = temp_de['Temperatur'].rolling(window=window_size, center=True).mean()

# Vertikale Linie bei 1980
jahr_1980 = pd.to_datetime('1980-01-01')

# Steigung/Mittelwert 1880-1980
maske_1 = (temp_de['Jahr'] >= 1880) & (temp_de['Jahr'] < 1980)
x1 = temp_de.loc[maske_1, 'Jahr']
y1 = temp_de.loc[maske_1, 'Rolling_Avg']
# Drop NaNs in y1 and use the same indices for x1
valid_idx_1 = y1.dropna().index
x1_valid = x1.loc[valid_idx_1]
y1_valid = y1.loc[valid_idx_1]
steigung_1 = np.polyfit(x1_valid, y1_valid, 1)[0]

# Steigung/Mittelwert 1980-2024
maske_2 = (temp_de['Jahr'] >= 1980) & (temp_de['Jahr'] <= 2024)
x2 = temp_de.loc[maske_2, 'Jahr']
y2 = temp_de.loc[maske_2, 'Rolling_Avg']
valid_idx_2 = y2.dropna().index
x2_valid = x2.loc[valid_idx_2]
y2_valid = y2.loc[valid_idx_2]
steigung_2 = np.polyfit(x2_valid, y2_valid, 1)[0]

plt.figure(figsize=(14,6))
plt.plot(temp_de['Datum'], temp_de['Temperatur'], label='Monatliche Temperatur', alpha=0.5)
plt.plot(temp_de['Datum'], temp_de['Rolling_Avg'], color='red', label=f'Rolling Avg ({window_size} Monate)')

# Trendlinie 1880-1980
plt.plot(
    pd.to_datetime(x1_valid, format='%Y'),
    np.polyval(np.polyfit(x1_valid, y1_valid, 1), x1_valid),
    color='blue', linewidth=2, label=f'Trend 1880-1980 (Steigung: {steigung_1:.4f}°C/Jahr)'
)
# Trendlinie 1980-2024
plt.plot(
    pd.to_datetime(x2_valid, format='%Y'),
    np.polyval(np.polyfit(x2_valid, y2_valid, 1), x2_valid),
    color='green', linewidth=2, label=f'Trend 1980-2024 (Steigung: {steigung_2:.4f}°C/Jahr)'
)

plt.axvline(jahr_1980, color='black', linestyle='--', label='1980')

plt.xlabel('Datum')
plt.ylabel('Temperatur (°C)')
plt.title('Gleitender Mittelwert der Temperatur in Deutschland mit Trendlinien')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Durchschnittstemperatur für Deutschland im Jahr 1980 berechnen
de_1980 = temp_hist[(temp_hist['Region'] == 'Deutschland') & (temp_hist['Jahr'] == 1980)]
durchschnitt_1980 = de_1980['Temperatur'].mean()
print(f"Durchschnittstemperatur für Deutschland im Jahr 1980: {durchschnitt_1980:.2f} °C")
```

2080 ist die jährliche Durchschnittstemperatur 12,09 °C, wenn die Steigung so bleibt. 

Steigen die Werte in den nächsten Jahren mehr als die Steigungsgerade annimmt, dann wird die Jahresdurchschnittstemperatur 2080 höher sein als 12,09 °C.
